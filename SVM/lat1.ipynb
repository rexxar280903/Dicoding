```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = np.array([
    [1, 2, 3, +1],
    [2, 3, 3, +1],
    [3, 1, 1, -1],
    [4, 2, 1, -1]
])

# Pisahkan kolom untuk scatter plot
x = data[:, 1]  # Ambil x1 (kolom kedua)
y = data[:, 2]  # Ambil x2 (kolom ketiga)
classes = data[:, 3]  # Kelas (+1 atau -1)

# Plot scatter
plt.scatter(x, y, c=classes, cmap='bwr', edgecolors='k')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter Plot Data")
plt.colorbar(label="Class")
plt.show()
```
